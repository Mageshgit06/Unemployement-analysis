```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("Unemployment in India.csv")
df2 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
from IPython.display import display
#DisplayDetails
display(df1.head())
display(df2.head())
```

```python
#RemoveUnwantedSpaces
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
```

```python
df1.info()
df2.info()
```

```python
df1.isnull().sum()
df2.isnull().sum()
```

```python
#RemovesNULLvalues
df1.dropna(inplace=True)
df2.dropna(inplace=True)
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])
```

```python
#Visualofdata1
plt.figure(figsize=(10,5))
plt.plot(df1['Date'], df1['Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Region", y="Estimated Unemployment Rate (%)", data=df1)
plt.xticks(rotation=90)
plt.show()
```

```python
#visualofdata2
sns.lineplot(x="Date", y="Estimated Unemployment Rate (%)", hue="Region", data=df2)
plt.show()
```

```python
df1.groupby("Region")["Estimated Unemployment Rate (%)"].mean().sort_values(ascending=False)
```
